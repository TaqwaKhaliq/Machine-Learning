```python
import numpy as np
```

```python
x = np.arange(1,101)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y, 'k--') #Color of the plot, line style
plt.xlabel('x')
plt.ylabel('sinx')
plt.title('Graph of Sin x')
```

```python
x=np.linspace(1,10,100)
y=x+2
plt.plot(x,y, 'k--') #Color of the plot, line style
plt.xlabel('x')
plt.ylabel('sinx')
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--')
plt.minorticks_on()
plt.title('Graph of Sin x')
```

```python
plt.plot(x,y,linestyle='dashed')
```

```python
#Specifying limits
plt.plot(x, np.tan(x)) 
plt.xlim(-1, 11) 
plt.ylim(-1.5, 1.5)
```

```python
#Subplots: Having more than one plots in a figure
```

```python
x = np.linspace(0, 2*np.pi, 400) #400 equidistant points between 0 and 2*pi
y = np.sin(x**2)
f, (ax1, ax2) = plt.subplots(1, 2) #Two figures in one row
ax1.plot(x, y)
ax1.set_title('Figure 1') #Setting title of first figure
ax2.scatter(x, y)
ax2.set_title('Figure 2') #Setting title of second figure
```
